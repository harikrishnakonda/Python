## Data Preprocessing - Encoding

```python
df.head()
```

```python
categorical_columns = ['sex','children','smoker','region']

# One hot encoding
df_encoded = pd.get_dummies(data = df,prefix='OHE',prefix_sep = '_',columns=categorical_columns,drop_first=True,dtype='int8')
```

```python
df_encoded.head()
df_encoded['charges'] = np.log10(df_encoded['charges'])
```

```python
df_encoded.head()
```

```python
X = df_encoded.drop(['charges'],axis=1) #input features
y = df_encoded['charges'] #target variable
```

```python
#np.random.seed(42)
np.random.randn(2,1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

## Training the model

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

## Test the Model

```python
y_pred = lin_reg.predict(X_test)
```

```python
y_pred
```

```python
mean_squared_error(y_pred,y_test) ##loss is 0.03
```

## R2 Square

```python
lin_reg.score(X_test,y_test) ## 0.779*100 - 77.9% accurate
```
