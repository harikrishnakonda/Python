```python
# we will use seaborn to actually plot the plot
# we will use matplotlib to have some settings
import matplotlib.pyplot as plt #This plots are not that visually good
import seaborn as sns #This plots the graphics of plot or colour it uses are more neater

## set the figure size

plt.figure(figsize=(10,5))

ax = sns.countplot(x='salary',data=df)
for p in ax.patches:
    ax.annotate((p.get_height()),(p.get_x()+0.3,p.get_height()+10),fontsize = 10)
```

```python
percentage = df['salary'].value_counts()/len(df) 
percentage
```

```python
# plt.pie(percentage,labels = ['<=50k','>50k'],autopct = '%.1f%%',explode=[0,0.1],
#        colors = ['lightskyblue','gold'],shadow=True)
df['salary'].value_counts().plot(kind='pie',autopct='%.1f%%',explode=[0,0.1],colors=['lightskyblue','gold'],shadow=True)
plt.title("Pie chart of salary",fontsize = 10)
```
